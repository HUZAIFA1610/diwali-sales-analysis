# Exploratory Data Analysis/ Data Visualization

### Gender

```python
# plotting a bar chart for Gender and it's count
sns.set(rc={'figure.figsize':(4,5)})
custom_colors = ['purple', 'brown']
ax = sns.countplot(x = 'Gender',data = df, palette=custom_colors, hue='Gender', dodge=False, legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, palette='viridis', hue='Gender', dodge=False, legend=False)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
sns.set(rc={'figure.figsize':(10,5)})
custom_colors = ["#FF6F61", "#6A0572"]
ax = sns.countplot(data = df, x = 'Age Group', palette=custom_colors, hue = 'Gender', legend=False)

# Add a legend to identify Male and Female
ax.legend(title='Gender', loc='upper right', labels=['Male', 'Female'])

# Add a title and labels for better readability
plt.title('Count of Age Groups by Gender', fontsize=16, weight='bold')
plt.xlabel('Age Group', fontsize=12, weight='bold')
plt.ylabel('Count', fontsize=12, weight='bold')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

#sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue='Age Group', dodge=False, palette='custom_colors', legend=False)
custom_colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink']
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette=custom_colors, hue='Age Group', dodge=False, legend=False)
plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(data = sales_state, x = 'State',y= 'Orders', hue='State', dodge=False, palette='viridis', legend=False)

# Add numerical annotations on top of each bar
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f', label_type='edge', fontsize=10, color='black')

# Add a title and labels for better readability
plt.title('Total Orders by State', fontsize=16, weight='bold')
plt.xlabel('State', fontsize=12, weight='bold')
plt.ylabel('Total Orders', fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```

```python
# total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

custom_colors = ["#FF6F61", "#6A0572", "#28B463", "#FFC300", "#1F77B4",
                 "#D62728", "#9467BD", "#8C564B", "#E377C2", "#2CA02C"]

# Set the figure size
sns.set(rc={'figure.figsize':(15,5)})

ax = sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue = 'State', dodge=False, palette=custom_colors, legend=False)

# Add annotations on top of the bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f', label_type='edge', fontsize=10, color='black')

# Add a title and axis labels
plt.title('Total Sales Amount from Top 10 States', fontsize=18, weight='bold')
plt.xlabel('State', fontsize=14, weight='bold')
plt.ylabel('Total Sales Amount', fontsize=12, weight='bold')

# Add gridlines for better structure
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
sns.set(rc={'figure.figsize': (4, 6)})

custom_colors = ['green', 'purple']
ax = sns.countplot(data = df, x = 'Marital_Status', palette = custom_colors, hue = 'Marital_Status', legend = False)

for bars in ax.containers:
    ax.bar_label(bars)

    # Add a legend to identify marital status
legend_labels = ['Single', 'Married']  # Update based on the unique values in your dataset
plt.legend(title='Marital Status', labels=legend_labels, loc='upper right', fontsize=10)

plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', palette='viridis', hue='Gender', legend=True)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', palette = 'viridis', hue = 'Occupation', dodge = False, legend = False)

for bars in ax.containers:
    ax.bar_label(bars)

# Add title and axis labels with bold font
ax.set_title('Total Order count by Occupation', fontsize=18, weight='bold', pad=20)
ax.set_xlabel('Occupation', fontsize=14, weight='bold')
ax.set_ylabel('Total Order Count', fontsize=14, weight='bold')

plt.show()
```

```python
# Create grouped data
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size and style
sns.set(rc={'figure.figsize': (20, 7)})
sns.set_style("whitegrid")

# Define custom vibrant colors
custom_colors = [
    "#FF6F61", "#6B5B95", "#88B04B", "#F7CAC9", "#92A8D1",
    "#955251", "#B565A7", "#009B77", "#DD4124", "#D65076",
    "#45B8AC", "#EFC050", "#5B5EA6", "#9B2335", "#DFCFBE"
]

# Create the barplot
ax = sns.barplot(
    data=sales_state, 
    x='Occupation', 
    y='Amount', 
    palette=custom_colors,
    hue = 'Occupation',
    legend = False
)

# Rotate x-axis labels for better readability
ax.tick_params(axis='x', rotation=45)

# Add data labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars, labels=[f'{int(value):,}' for value in bars.datavalues], fontsize=10, color='black', weight='bold')

# Add title and axis labels with bold font
ax.set_title('Total Sales Amount by Occupation', fontsize=18, weight='bold', pad=20)
ax.set_xlabel('Occupation', fontsize=14, weight='bold')
ax.set_ylabel('Total Sales Amount', fontsize=14, weight='bold')

# Add a grid for structure
ax.grid(color='lightgrey', linestyle='--', linewidth=0.5, axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,7)})
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category', palette = custom_colors, legend = False)

#Define custom colors
custom_colors = [
    "#FF5733", "#33FF57", "#3357FF", "#FF33A1", "#FFC300",
    "#DAF7A6", "#581845", "#900C3F", "#C70039", "#FF6F61",
    "#A6052A", "#1F618D", "#2B4863", "#F39C12", "#D35400",
    "#5DADE2", "#AF7AC5", "#E74C3C", "#76D7C4", "#F1948A"
]

# Rotate x-axis labels for better readability
ax.tick_params(axis='x', rotation=45)

# Add title and axis labels with bold font
ax.set_title('Total Sales by Product Category', fontsize=18, weight='bold', pad=20)
ax.set_xlabel('Product Category', fontsize=14, weight='bold')
ax.set_ylabel('Total Sales', fontsize=14, weight='bold')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size and style
sns.set(rc={'figure.figsize': (15, 5)})
sns.set_style("whitegrid")

custom_colors = [
    "#581845", "#6A0572", "#28B463", "#FF33A1", "#FFC300",
    "#3357FF", "#FF5733", "#900C3F", "#C70039", "#FF6F61"
]

ax = sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', palette = custom_colors, hue = 'Product_Category', legend = False)

# Rotate x-axis labels for better readability
ax.tick_params(axis='x', rotation=45)

# Add title and axis labels with bold font
ax.set_title('Total Sales Amount by Product Category', fontsize=18, weight='bold', pad=20)
ax.set_xlabel('Product Category', fontsize=14, weight='bold')
ax.set_ylabel('Total Sales Amount', fontsize=14, weight='bold')

# Add data labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars, labels=[f'{int(value):,}' for value in bars.datavalues], fontsize=10, color='black', weight='bold')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
